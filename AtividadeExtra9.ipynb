### Resolução dos Exemplos de Memória Cache (5.2 a 5.7)
Este notebook contém as implementações em Python para os cálculos de capacidade, tags e formatos de endereço de memória cache conforme as transcrições fornecidas.

```python
import math

def calc_log2(value):
    return int(math.log2(value))

# --- Exemplo 5.2: Cálculo de bits totais (Mapeamento Direto) ---
print("--- Exemplo 5.2 ---")
cache_cap_kb = 32
largura_linha_b = 8
mp_cap_mb = 16

td_bits = cache_cap_kb * 1024 * 8
num_linhas = (cache_cap_kb * 1024) / largura_linha_b
num_blocos_mp = (mp_cap_mb * 1024 * 1024) / largura_linha_b

# Tag = Blocos / Linhas
largura_tag = calc_log2(num_blocos_mp / num_linhas)
tt_bits = num_linhas * largura_tag
t_total = td_bits + tt_bits

print(f"Total de bits (Td): {td_bits}")
print(f"Largura do Tag: {largura_tag} bits")
print(f"Total de bits da Cache (T): {t_total} bits")
```

```python
# --- Exemplo 5.3: Formato de Endereço (Mapeamento Direto) ---
print("\n--- Exemplo 5.3 ---")
mp_64mb = 64 * 1024 * 1024
cache_2k_linhas = 2 * 1024
largura_16b = 16

bits_endereco_total = calc_log2(mp_64mb)
bits_byte = calc_log2(largura_16b)
bits_linha = calc_log2(cache_2k_linhas)
bits_tag = bits_endereco_total - bits_linha - bits_byte

print(f"Endereço Total: {bits_endereco_total} bits")
print(f"Campos: Tag({bits_tag}) | Linha({bits_linha}) | Byte({bits_byte})")
```

```python
# --- Exemplo 5.4: Localização de Linha (Hex -> Bin) ---
print("\n--- Exemplo 5.4 ---")
endereco_hex = "3FC92B6"
largura_bloco = 32
cache_cap = 128 * 1024

bits_total = 28 # 7 hex * 4
bits_byte = calc_log2(largura_bloco)
num_linhas = cache_cap / largura_bloco
bits_linha = calc_log2(num_linhas)

# Converter hex para binário com preenchimento de zeros (28 bits)
endereco_bin = bin(int(endereco_hex, 16))[2:].zfill(bits_total)

# Divisão dos campos: Tag | Linha | Byte
# Tag = total - linha - byte | Linha | Byte
tag_part = endereco_bin[:(bits_total - bits_linha - bits_byte)]
linha_part = endereco_bin[(bits_total - bits_linha - bits_byte):-bits_byte]
byte_part = endereco_bin[-bits_byte:]

print(f"Endereço Binário: {endereco_bin}")
print(f"Valor binário da Linha: {linha_part}")
```

```python
# --- Exemplo 5.6: Mapeamento Totalmente Associativo ---
print("\n--- Exemplo 5.6 ---")
mp_64mb = 64 * 1024 * 1024
largura_16b = 16

bits_total = calc_log2(mp_64mb)
bits_byte = calc_log2(largura_16b)
bits_bloco = bits_total - bits_byte

print(f"Formato Associativo: Bloco({bits_bloco} bits) | Byte({bits_byte} bits)")
```

```python
# --- Exemplo 5.5: Cálculo de bits totais (Mapeamento Associativo) ---
print("\n--- Exemplo 5.5 ---")
cache_cap_kb = 32
largura_linha_b = 8
mp_cap_mb = 16

td_bits = cache_cap_kb * 1024 * 8
num_linhas = (cache_cap_kb * 1024) / largura_linha_b
num_blocos_mp = (mp_cap_mb * 1024 * 1024) / largura_linha_b

# No associativo completo, o campo 'bloco' identifica o bloco da MP
largura_campo_bloco = calc_log2(num_blocos_mp)
tb_bits = num_linhas * largura_campo_bloco
t_total_assoc = td_bits + tb_bits

print(f"Total de bits de dados (Td): {td_bits}")
print(f"Largura do campo Bloco: {largura_campo_bloco} bits")
print(f"Total de bits da Cache: {t_total_assoc} bits")
```

```python
# --- Exemplo 5.7: Valor do Campo Bloco (Hex -> Bin) ---
print("\n--- Exemplo 5.7 ---")
endereco_hex = "3FC92B6"
largura_bloco_b = 32

bits_total = 28
bits_byte = calc_log2(largura_bloco_b)
bits_bloco = bits_total - bits_byte

endereco_bin = bin(int(endereco_hex, 16))[2:].zfill(bits_total)

# No associativo, o campo bloco são os bits mais significativos antes do byte
bloco_part = endereco_bin[:bits_bloco]

print(f"Endereço Binário: {endereco_bin}")
print(f"Valor binário do campo Bloco: {bloco_part}")
```
